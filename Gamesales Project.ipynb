```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
game_sales = pd.read_csv("Gamesales.csv")
game_sales.head()
```

```python
game_sales.info()
```

```python
game_sales.describe()
```

```python
##cleaning data or null values

print(game_sales.isnull().sum())
##

```

```python
game_cleaned = game_sales.dropna()



```

```python
##check duplicates
print(game_cleaned.duplicated().sum())

```

```python
game_cleaned
```

```python
#we have to make copy of subset beacause if we change directly it will also change orignal df  
game_cleaned = game_cleaned.copy()
```

```python
game_cleaned["Year"] = game_cleaned["Year"].astype(int)
```

```python
game_cleaned
```

```python
game_cleaned.info()
```

```python
game_cleaned.describe()
```

```python
game_cleaned.head(5)
```

```python
fig , ax = plt.subplots()
ax.bar(game_cleaned["Year"] , game_cleaned["Global_Sales"]);
plt.show();
```

```python
fig ,ax1= plt.subplots()
#global sales by 



ax1 = game_cleaned.groupby("Year")["Global_Sales"].sum().plot(kind ="line" ,marker = "o");
plt.show();
```

```python
## bar plot of global sales by genre
fig, ax2 = plt.subplots()
ax2 =  game_cleaned.groupby("Genre")["Global_Sales"].sum().plot(kind = "bar")
plt.show()
```

## which game  are played most

```python
#which platfrom has highest sales
fig,ax3 = plt.subplots()
a3 = game_cleaned.groupby("Platform")["Global_Sales"].sum().plot(kind = "bar");
a3.set(ylabel = "Sales");
plt.show();
```

## which game  are played most


```python
## sales by regions
fig , ax4 = plt.subplots()
ax4 = game_cleaned.groupby("Year")[["NA_Sales" , "EU_Sales","JP_Sales"]].sum().plot(kind="bar");
plt.legend(title = "Regions");
plt.show();
```

```python
game_cleaned.corr(numeric_only=True)
```

## which game  are played most

```python
#which game  are played most
genre_counts = game_cleaned["Genre"].value_counts()
fig ,ax6 = plt.subplots(figsize = (12,7))
ax6.bar(genre_counts.index , genre_counts.values);
plt.show()
```

## which platform has launched most game


```python
#which platform has launched most game 
#ps2 has launched most game so it has highest sales
genre_counts = game_cleaned["Platform"].value_counts()
fig ,ax7 = plt.subplots(figsize = (30,15))
ax7.bar(genre_counts.index , genre_counts.values);
plt.show()
```

## which year games was as its highest

```python
year_count = game_cleaned["Year"].value_counts()
fig , ax8  = plt.subplots()
ax8.bar(year_count.index , year_count.values)
plt.show()
## as you can see most games was launches at 2008 and 2009
```
